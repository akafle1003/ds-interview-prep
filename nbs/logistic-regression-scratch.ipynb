```python
import numpy as np
import sklearn.datasets

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def loss(p, y):
    return np.mean(-y * np.log(p) - (1 - y) * np.log(1 - p))


def gradient(p, X, y):
    return np.dot(X.T, (p -y)) / y.shape[0]
```

```python
# initialize variables
iris = sklearn.datasets.load_iris()

# choose 2 features
X = iris.data[:, :2]
y = np.array((iris.target != 0) * 1)
```

```python
lr = 0.001
theta = np.zeros(X.shape[1])
```

```python
theta
```

```python
z = np.dot(X, theta)
p = sigmoid(z)
```

```python
lz = 1
i = 0
while lz > 0.05:
    z = np.dot(X, theta)
    p = sigmoid(z)
    
    if i % 10000 == 0:
        lz = loss(p, y)
        print(lz)

    grad = np.dot(X.T, p-y) / y.shape[0]
    theta -= lr * grad
    
    i += 1
```

```python
pred = np.round(sigmoid(np.dot(X, theta)),2)
compare = np.zeros(shape=(X.shape[0],2))
compare[:,0] = y
compare[:,1] = pred
compare
```
