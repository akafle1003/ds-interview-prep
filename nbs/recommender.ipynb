```python
import pandas as pd
import numpy as np
import os
```

```python
def load_data():
    df = pd.read_csv("../test_data/recommender/file.tsv",
                     sep='\t',
                     header=None)

    df.columns = ['user_id', 'item_id', 'rating', 'timestamp']

    titles = pd.read_csv("../test_data/recommender/Movie_Id_Titles.csv")
    df = df.merge(titles, how='left', on='item_id')
    return df, titles
```

```python
df, titles = load_data()
df.head()
```

### Average rating

```python
summary_ratings = df.groupby('title')[['rating']].agg(['mean', 'count'])
summary_ratings.columns = ['mean_rating', 'rating_count']
```

### Correlation Matrix Approach

```python
piv_table = df.pivot_table(values='rating', index='user_id', columns='title')
correlation_matrix = piv_table.corr()
```

```python
starwars_user_ratings = piv_table['Star Wars (1977)']
best_matches = pd.DataFrame(piv_table.corrwith(starwars_user_ratings).sort_values(ascending=False))
merged = best_matches.merge(summary_ratings, how='left', left_index=True, right_index=True)
merged[merged['rating_count'] > 100].head(20)
```
